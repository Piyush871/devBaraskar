Data Cleaning

```python
#Delete 'region','city','state_code' as they provide too much of granularity.
columns1=['region','city','state_code']
df.drop(columns=columns1,inplace=True)
```

```python
#Delete 'id', 'Unnamed: 0.1', 'entity_type', 'entity_id', 'parent_id', 'created_by','created_at', 'updated_at' as they are redundant.
columns2=['id','Unnamed: 0.1', 'entity_type', 'entity_id', 'parent_id', 'created_by',
   'created_at', 'updated_at']
df.drop(columns=columns2,inplace=True)
```

```python
# Delete 'domain', 'homepage_url', 'twitter_username', 'logo_url', 'logo_width', 'logo_height',            'short_description', 'description', 'overview','tag_list', 'name', 'normalized_name', 'permalink',    'invested_companies' as they are irrelevant features.
columns3=['domain', 'homepage_url', 'twitter_username', 'logo_url', 'logo_width', 'logo_height',
    'short_description', 'description', 'overview','tag_list', 'name', 'normalized_name', 'permalink',
    'invested_companies']
df.drop(columns=columns3,inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
threshold = 0.98
columns_to_drop = df.columns[df.isnull().mean() > threshold]
df.drop(columns=columns_to_drop, inplace=True)
```

```python
#Delete instances with missing values for 'status', 'country_code', 'category_code' and 'founded_at'.
df.dropna(subset=['status','country_code','category_code','founded_at'],inplace=True)
```

```python
df.isnull().sum()
```

```python
# Scatter plot
fig, ax = plt.subplots(figsize=(6, 4))
ax.scatter(df['funding_total_usd'], df['funding_rounds'])

# x-axis label
ax.set_xlabel('funding_total_usd')

# y-axis label
ax.set_ylabel('funding_rounds')
plt.show()
```

```python
for x in ['funding_total_usd']:
    q75,q25 = np.percentile(df.loc[:,x],[75,25])
    intr_qr = q75-q25

    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)

    df.loc[df[x] < min,x] = np.nan
    df.loc[df[x] > max,x] = np.nan
```

```python
for x in ['funding_rounds']:
    q75,q25 = np.percentile(df.loc[:,x],[75,25])
    intr_qr = q75-q25

    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)

    df.loc[df[x] < min,x] = np.nan
    df.loc[df[x] > max,x] = np.nan
```

```python
df.dropna(subset=['funding_total_usd','funding_rounds','status'],inplace=True)
```

Date Transformation

```python
#founded_at
df['founded_at'] = df['founded_at'].fillna('2021')
df['founded_at'] = df['founded_at'].astype(str)
df['founded_at'] = df['founded_at'].str[0:4].astype(int)

	#closed at
df['closed_at'] = df['closed_at'].fillna('2021')
df['closed_at'] = df['closed_at'].astype(str)
df['closed_at'] = df['closed_at'].str[0:4].astype(int)

#first_funding_at
df['first_funding_at'] = df['first_funding_at'].fillna('2021')
df['first_funding_at'] = df['first_funding_at'].astype(str)
df['first_funding_at'] = df['first_funding_at'].str[0:4].astype(int)


#last_funding_at
df['last_funding_at'] = df['last_funding_at'].fillna('2021')
df['last_funding_at'] = df['last_funding_at'].astype(str)
df['last_funding_at'] = df['last_funding_at'].str[0:4].astype(int)


#first_milestone_at
df['first_milestone_at'] = df['first_milestone_at'].fillna('2021')
df['first_milestone_at'] = df['first_milestone_at'].astype(str)
df['first_milestone_at'] = df['first_milestone_at'].str[0:4].astype(int)

#last_milestone_at
df['last_milestone_at'] = df['last_milestone_at'].fillna('2021')
df['last_milestone_at'] = df['last_milestone_at'].astype(str)
df['last_milestone_at'] = df['last_milestone_at'].str[0:4].astype(int)
```

creating new variables

```python
# a. Create new feature 'isClosed'
df['isclosed'] = (df['status'] == 'closed') | (~df['closed_at'].isnull())
```

```python
isClosed=[]

for index,row in df.iterrows():
  if row['status'] == 'operating' or row['status'] == 'ipo' :
    result=1
    isClosed.append(result)
  elif row['status'] =='acquired' or row['status'] =='closed' :
    result2=0
    isClosed.append(result2)

df.loc[:, 'isclosed'] = isClosed
```

```python
# Relationship between 'founded_at' and 'status'
plt.figure(figsize=(12, 6))
sns.boxplot(x='status', y='founded_at', data=df)
plt.title('Relationship between Founding Year and Startup Status')
plt.show()
```
